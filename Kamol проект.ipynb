Исходя из этой информации, использовал методы fillna() и dropna() для обработки пропущенных значений в каждом столбце.

```python
import pandas as pd

data = pd.read_csv('/Users/kamol/Downloads/ch_books.csv')
data = data.dropna(subset=['title'])
data['author'] = data['author'].fillna('Unknown')
data = data.dropna(subset=['annotation'])
data['publisher'] = data['publisher'].fillna('Unknown')
data['series'] = data['series'].fillna('')
data['year'] = data['year'].fillna(data['year'].mean())
data = data.dropna(subset=['isbn', 'circulation'])

numeric_columns = ['pages', 'height', 'width', 'thickness', 'weight', 'price']
data[numeric_columns] = data[numeric_columns].fillna(data[numeric_columns].mean())

categorical_columns = ['cover_type', 'age_restrictions']
data[categorical_columns] = data[categorical_columns].fillna(data[categorical_columns].mode().iloc[0])

missing_values_after = data.isnull().sum()
print("Пропущенные значения после обработки:\n", missing_values_after)
```

# Визуализация данных:

```python
import matplotlib.pyplot as plt
genre_counts = data['genre_name'].value_counts()

plt.figure(figsize=(15, 6))
plt.bar(genre_counts.index, genre_counts.values)
plt.xlabel('Жанр')
plt.ylabel('Количество книг')
plt.title('Распределение книг по жанрам')
plt.xticks(rotation=45)
plt.show()
```

```python
cover_type_counts = data['cover_type'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(cover_type_counts.values, labels=cover_type_counts.index, autopct='%1.1f%%')
plt.title('Соотношение типов обложек')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['pages'], data['price'])
plt.xlabel('Количество страниц')
plt.ylabel('Цена')
plt.title('Связь между количеством страниц и ценой книг')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['price'], bins=100, edgecolor='black')
plt.xlabel('Цена')
plt.ylabel('Частота')
plt.title('Распределение цены книг')
plt.show()
```


# Гипотеза

Гипотеза: Книги с более высоким рейтингом имеют более высокую цену.

Реализация: Для проверки этой гипотезы можно использовать коэффициент корреляции, например, коэффициент корреляции Пирсона, для измерения связи между рейтингом и ценой книг. Другой вариант - разделить книги на группы с высоким и низким рейтингом и сравнить среднюю цену в каждой группе с помощью теста Стьюдента или непараметрического теста, например, критерия Манна-Уитни.

Гипотеза: Книги с большим числом страниц стоят дороже, чем книги с меньшим числом страниц.

Реализация: Для проверки этой гипотезы можно разделить книги на группы с большим и меньшим количеством страниц и сравнить средние цены в каждой группе с помощью теста Стьюдента или непараметрического теста, например, критерия Уилкоксона.
